# Create and save vocabulary

```python
vocab_store_id = np.array(df_sales_valid.store_id.drop_duplicates().sort_values())
```

```python
np.save('vocab_store_id.npy', vocab_store_id, allow_pickle=True)
```

```python
vocab_item_id = np.array(df_sales_valid.item_id.drop_duplicates().sort_values())
np.save('vocab_item_id.npy', vocab_item_id, allow_pickle=True)
```

```python
vocab_dept_id = np.array(df_sales_valid.dept_id.drop_duplicates().sort_values())
np.save('vocab_dept_id.npy', vocab_dept_id)
```

```python
vocab_state_id = np.array(df_sales_valid.state_id.drop_duplicates().sort_values())
np.save('vocab_state_id.npy', vocab_state_id, allow_pickle=True)
```

```python
vocab_cat_id = np.array(df_sales_valid.cat_id.drop_duplicates().sort_values())
np.save('vocab_cat_id.npy', vocab_cat_id, allow_pickle=True)
```
